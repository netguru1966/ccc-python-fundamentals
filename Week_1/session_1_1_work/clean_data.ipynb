Cleaning Data

```python
# Messy data from spreadsheet
raw_data = [
        ' JOHN  SMITH ',
        ' alice  JOHNSON',
        'BOB  jones',
        ' DArth VadER ' 
]
```

```python
# Clean it up !
clean_data = []
for name in raw_data:
    #Remove spaces, makes title case
    clean = name.strip().title()
    clean_data.append(clean)
print(clean_data)
```

```python
# ['John Smith', 'Alice Johnson', 'Bob Jones']
# You just processed data professionally
```

```python
# Expanded messy data (including extra middle spaces and mixed types)
raw_people = [
    {'name': '  JOHN  SMITH  ', 'city': 'new  york'},
    {'name': 'alice  JOHNSON', 'city': 'LOS ANGELES'},
    {'name': 'BOB  jones', 'city': 'chicago'},
    {'name': ' DArth VadER ', 'city': 'DEATH STAR'}
]
```

```python
clean_people = []
```

```python
for person in raw_people:
    # Use split() and join() to remove ALL extra whitespace (inner and outer)
    clean_name = " ".join(person['name'].split()).title()
    clean_city = " ".join(person['city'].split()).title()
    
    clean_people.append({
        'name': clean_name,
        'city': clean_city
    })
```

```python
# Preview the professional results
for entry in clean_people:
    print(f"Name: {entry['name']:15} | City: {entry['city']}")
```

Split and Join - Processing Text

```python
sentence = 'I love learning Python for ML'
```

```python
# Split into words (creates a list)
words = sentence.split()
print(words)
```

```python
# Count words
word_count = len(words)
print(f'Word count: {word_count}') # 6
```

```python
# Join words back together
rejoined = ' '.join(words)
print(rejoined)
```
