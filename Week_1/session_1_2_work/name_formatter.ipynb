```python
# Exercise 1: Name Formatter
name1 = 'john smith'
name2 = '  ALICE JOHNSON  '
name3 = 'bob   JONES'

# This is my MOD
```

```python
formatted_name1 = name1.strip().title()
formatted_name2 = name2.strip().title()
formatted_name3 = name3.strip().title()
```

```python
print("=== Name Formatter ===\n")
print("Formatted Names:")
print(f"1. {formatted_name1}")
print(f"2. {formatted_name2}")
print(f"3. {formatted_name3}")
```
