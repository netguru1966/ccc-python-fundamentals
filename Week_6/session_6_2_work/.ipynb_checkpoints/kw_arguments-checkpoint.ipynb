## # Function accepts any keyword arguments

```python
def create_profile(**kwargs):
    print('Profile created:')
    for key, value in kwargs.items():
        print(f'  {key}: {value}')
```

```python
# Call with different fields

create_profile(name='Vanay', age=59)

# Profile created:
#   name: Vanay
#   age: 59
```

```python
create_profile(name='Bob', age=30, city='NYC', job='Engineer')

# Profile created:
#   name: Bob
#   age: 30
#   city: NYC
#   job: Engineer
```

```python
# Flexible key-value pairs!
```

```python
create_profile(name='Vanay', age=59, city='Chicago', job='Cybersecurity')
```
