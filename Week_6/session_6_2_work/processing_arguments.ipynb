## Processing *args

### Find MAX

```python
# *args is a tuple - loop through it

def find_max(*args):
    if not args:
        return None
    max_val = args[0]
    for num in args:
        if num > max_val:
            max_val = num
    return max_val
```

```python
print(find_max(5, 2, 9, 1, 7))    # 9
print(find_max(100, 50, 75))      # 100

# Or just use max():

def find_max(*args):
    return max(args) if args else None
```

### Find MIN

```python
def find_min(*args):
    if not args:
        return None
    min_val = args[0]
    for num in args:
        if num < min_val:
            min_val = num
    return min_val
```

```python
print(find_min(5, 2, 9, 1, 7))    # 9
print(find_min(100, 50, 75))      # 100
```

```python
# Or just use min():

def find_min(*args):
    return min(args) if args else None
```
