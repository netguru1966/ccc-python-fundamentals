## Simple operations

```python
# Simple operations
add = lambda a, b: a + b
print(add(10, 5))      # 15
```

```python
# String operations
uppercase = lambda s: s.upper()
print(uppercase('cyber'))   # HELLO
```

```python
# Multiple arguments
full_name = lambda first, last: f'{first} {last}'
print(full_name('Vanay', 'Henry'))   # Alice Smith
```

```python
# Conditional

is_adult = lambda age: age >= 18
print(is_adult(25))    # True
print(is_adult(15))    # False

# Quick and clean!
```
