## Define function with parameter

```python
def greet(name):
    print(f'Hello, {name}!')
    print('Welcome to Python!')
```

```python
# Call with different arguments

greet('Vanay')

# Hello, Vanay!
# Welcome to Python!
```

```python
greet('Ameer')

# Hello, Ameer!
# Welcome to Python!
```

```python
greet('Maymuna')

# Hello, Maymuna!
# Welcome to Python!
```
