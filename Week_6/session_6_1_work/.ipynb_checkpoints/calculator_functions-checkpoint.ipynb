```python
def add(a, b): 
    return a + b

def subtract(a, b): 
    return a - b

def multiply(a, b): 
    return a * b

def divide(a, b):
    if b == 0:
        # Using raise to throw an exception instead of break
        # break can only be used inside loops
        raise ValueError("Cannot divide by zero")
    return a / b
```

```python
# Test each function:

print(add(10, 5))       # 15
print(subtract(10, 5))  # 5
print(multiply(10, 5))  # 50
print(divide(10, 5))    # 2.0
```
