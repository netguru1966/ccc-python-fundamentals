# Multiple Default Parameters

## Multiple defaults

```python
def create_user(name, age=50, city='Unknown'):
    print(f'Name: {name}')
    print(f'Age: {age}')
    print(f'City: {city}')
```

```python
# def create_user(name, age=50, city='Unknown'):
#    print(f'Name: {name} Age: {age} City: {city}')
```

```python
# Different ways to call:

create_user('Vanay')

# Name: Vanay, Age: 50, City: Unknown
```

```python
create_user('Bob', 25)

# Name: Bob, Age: 25, City: Unknown
```

```python
create_user('Triniti', 30, 'Chicago')

# Name: Vanay, Age: 30, City: Chicago
# Mix required and optional!
# Required first, defaults after
```
