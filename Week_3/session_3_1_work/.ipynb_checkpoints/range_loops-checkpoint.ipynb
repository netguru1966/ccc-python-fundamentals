```python
# Count to 5
for i in range(5): # This starts a for-loop
    print(i)
```

```python
# Count 1 to 10
for i in range(1, 11):
    print(i)
```

```python
# Count by 2's (even numbers)
for i in range(0, 11, 2):
    print(f'Output: {i}')
```

```python
# Count by 2's (odd numbers)
for i in range(1, 11, 2):
    print(i)
```

```python
# Count backwards!
for i in range(10, -1, -1):
 print(i)
```
