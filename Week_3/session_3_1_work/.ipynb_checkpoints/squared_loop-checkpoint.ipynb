```python
# Process each number
numbers = [1, 2, 3, 4, 5]
```

```python
for num in numbers:
    squared = num ** 2
    print(f'{num} squared is {squared}')
```

```python
# Output:
# 1 squared is 1
# 2 squared is 4
# 3 squared is 9
# 4 squared is 16
# 5 squared is 25
```

```python
# New Way

squared = [ num ** 2 for num in numbers]
print(f'{num} squared is {squared}')
```
