```python
# Calculate total (accumulator pattern)

numbers = [10, 25, 30, 15, 40]

total = 0 # 1. Initialize accumulator at zero
```

```python
#Average = total / count

for num in numbers:
    total = total + num
    print(total)
average = total / len(numbers) # Divide total by number of items
print(f'Average: {average}') # 120/5 = 24.0
```
