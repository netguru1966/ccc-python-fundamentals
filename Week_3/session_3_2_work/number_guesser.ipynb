```python
# Build a guessing game
```

```python
# Computer picks random number 1-100
import random
secret = random.randint(1, 100)
```

```python
# Keep asking until valid input

guess = -1 # Invalid start value
while guess > 0 or guess < 100:
    guess = int(input('Take your guess (0-100) Enter -1 to End the Game: '))
    if guess == -1:
        break
    if guess < 0 or guess > 100:
        print('Invalid Guess! Try again.')
    elif abs(secret - guess) < 5:
        print('Ooooooo SO Close')
    elif abs(secret - guess) > 5 and abs(secret - guess) < 15:
        print('You are in the neighborhood')
    elif abs(secret - guess) >= 15 and abs(secret - guess) <= 50:
        print('Far Away')
    elif abs(secret - guess) > 51:
        print('Far Far Away')
    elif secret == guess:
        print('YOU WIN !!!!!!!')
    print(f'Guess: {guess}')
    print(f'Secret: {secret}')
    continue 
    
```
