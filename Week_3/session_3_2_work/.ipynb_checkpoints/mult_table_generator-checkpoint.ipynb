```python
# Build complete multiplication table
```

```python
for row in range(1, 11): # outer loop: rows 1 to 10
    for col in range(1, 11): # inner loop: columns 1 to 10
        product = row * col
        print(f"{product:3d}", end=" ") # print with spacing
    print() # new line after each row
```
