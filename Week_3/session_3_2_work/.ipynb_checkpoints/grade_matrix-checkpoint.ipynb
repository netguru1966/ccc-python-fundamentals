```python
# Student grades: [student][subject]
# Grading Matrix
```

```python
grades = [[95,87,92,88],
[88,90,85,92],
[92,88,94,90], 
[78,85,80,88]
]
```

```python
# Calculate average for each student
for student_num, student_grades in enumerate(grades, 1):
    total = 0
    for grade in student_grades:
        total += grade
        avg = total / len(student_grades)
    print(f'Student {student_num} average: {avg:.1f}')
```

```python
# Calculate average for each subject
grades_transposed = list(zip(*grades))  # Transpose the grades matrix to get subjects as rows

for subject_num, subject_grades in enumerate(grades_transposed, 1):
    total = sum(subject_grades)  # Sum all grades for this subject
    avg = total / len(subject_grades)  # Calculate average
    print(f'Subject {subject_num} average: {avg:.1f}')
```

```python
# Overall Average
overall_average = total / len(grades)
print(f'Overall Average: {overall_average:.1f}')
```

```python
# Maximum grade within the class
max_grade = max(max(grades))
print(f'Maximum: {max_grade:.1f}')
```

```python
# How many A's total (>= 90)
count = 0
for row in grades:
    for grade in row:
        if grade >= 90:
            count += 1
            print(f'Grades over 90% : {grade}')
print(f'Total count: {count}')
```
