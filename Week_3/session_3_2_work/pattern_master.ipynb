```python
# Pattern 1: Square

for i in range(1, 6): # Row number
    for j in range(1,6): # Stars in this row
        print('*', end='') # end='' prevents newline
    print() # New line after each row
```

```python
# Pattern 2: Countdown

for i in range(6, 0, -1): # Row number
    for j in range(5, 6-i, -1): # Column number
        print(j,end=' ') # end='' prevents newline
    print() # New line after each row
```

This code will print:
- First row: 5 4 3 2 1
- Second row: 5 4 3 2
- Third row: 5 4 3
- Fourth row: 5 4
- Fifth row: 5
- Sixth row: (empty)

```python
# Pattern 3: Number grid (1-25 in 5Ã—5)

for row in range(1, 6): # outer loop: rows 1 to 5
    for col in range(1, 6): # inner loop: columns 1 to 5
        product = row * col
        print(f"{product:3d}", end=" ") # print with spacing
    print() # new line after each row
```
