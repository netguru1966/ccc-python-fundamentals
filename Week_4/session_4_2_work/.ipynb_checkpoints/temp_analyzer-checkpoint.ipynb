```python
# Given temperature data (Celsius):

temps = [22, -5, 18, 30, 15, -2, 25, 28, 12, 35, -10, 20]

# Use loops + conditionals to:

# 1. Count freezing days (< 0)
# 2. Count cold days (0–15)
# 3. Count comfortable days (16–25)
# 4. Count hot days (> 25)
# 5. Create separate list of valid temps (>= -50 and <= 50)
# 6. Calculate average of valid temps only
```

```python
temp_list = []
h_count = 0
cc_count = 0
c_count = 0
f_count = 0

for temp in temps:
    if temp >= -50 and temp <= 50:
        temp_list.append(temp)
        
    if temp > 25:
        h_count =+ 1
    elif temp > 16 and temp < 25:
        cc_count += 1
    elif temp > 0 and temp < 15:
        c_count += 1
    elif temp < 0:
        f_count += 1

average = sum(temp_list) / len(temp_list)
   
print(f'Temperature: {temp_list}')

print(f'Number of Hot Days in Celsuis: {h_count}')
print(f'Number of Comfortable Days in Celsuis: {cc_count}')
print(f'Number of Cold Days in Celsuis: {c_count}')
print(f'Number of Freezing Days in Celsuis: {f_count}')

print(f'Average: {average:.2f}')


```
