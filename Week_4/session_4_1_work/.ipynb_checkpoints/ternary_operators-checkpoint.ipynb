```python
# 1. Grade status in one line

score = 85
status = 'PASS' if score >= 60 else 'FAIL'
print(status)  # PASS
```

```python
# 2. Age category

age = 25
category = 'adult' if age >= 18 else 'minor'
print(category)
```

```python
# 3. Even/odd

num = 7
parity = 'even' if num % 2 == 0 else 'odd'
print(parity)
```

```python
# 4. Max of two numbers

a, b = 10, 15
max_value = a if a > b else b
print(max_value)
```
