```python
# Reverse a boolean value

is_raining = False
```

```python
# If NOT raining, go outside

if not is_raining:
    print('Go for a walk!')
else:
    print('Stay inside')
```

```python
# not reverses the value:
# not True -> False
# not False -> True
```

```python
# Another example

is_member = True

if not is_member:
    print('Please sign up')
```
