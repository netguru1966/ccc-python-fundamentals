```python
numbers = [42, 17, 93, 8, 55]
```

```python
# How many score
count = len(numbers)
print(f'Number of count: {count}')
```

```python
# Total count
total = sum(numbers)
print(f'Total count: {total}')
```

```python
# Average count
average = total / count
print(f'Average: {average:.2f}')
```

```python
# Hoghest and Lowest
print(f'Highest: {max(numbers)}')
print(f'Lowest: {min(numbers)}')
```
