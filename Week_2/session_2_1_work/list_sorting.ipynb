```python
numbers = [42, 17, 93, 8, 55]
```

```python
# Sort in ascending order
numbers.sort()
print(numbers)
```

```python
# Reverse in order
numbers.reverse()
print(numbers)
```

```python
# Name sort alphabetically
names = ['Charlie', 'Alice', 'Bob']
names.sort()
print(names)
```
