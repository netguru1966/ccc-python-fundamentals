```python
# Shopping List
```

```python
items = ['apples', 'bread', 'milk']
```

```python
# Add eggs to the list
items.append('eggs')
print(items)
```

```python
# Add cheese to the list
items.append('cheese')
print(items)
```

```python
# Remove bread from the list
items.remove('bread')
print(items)
```

```python
# Insert 'butter' at position 0
items.insert(0, 'butter')
print(items)
```

```python
# How many items in the shopping list
count = len(items)
print(f'Number of Items: {count}')
```

```python
# Get the first item in the list
print(items[0]) 
```

```python
# Get the last item in the list
print(items[-1]) 
```
