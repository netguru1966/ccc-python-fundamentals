```python
# List of Students
```

```python
students = ['Alice', 'Bob','Carol','David','Eve']
```

```python
# Get the first student
print(students[0]) #Alice
```

```python
# Get the second student
print(students[1]) #Bob
```

```python
# Get the last student
print(students[-1]) #Eve
```

```python
# Get the second to last student
print(students[-2]) #David
```
