```python
# Analyze weekly temperatures:
temperatures = [72, 68, 75, 70, 73, 69, 71]
```

```python
# How many temperatures in the list
count = len(temperatures)
print(f'Number of count: {count}')

# Total count
total = sum(temperatures)
print(f'Total count: {total}')

# Average count
average =  total / count
print(f'Average: {average:.2f}')
```

```python
# Highest and Lowest Tempertures
print(f'Highest: {max(temperatures)}')
print(f'Lowest: {min(temperatures)}')
```

```python
# Range between Highest and Lowest Tempertures
range = max(temperatures) - min(temperatures)
print(f'Range: {range}')
```

```python
# Iterate through the list

for temp in temperatures:
    if average > temp:
        print(f" Temperature above average {temp}.")
    if temp > 70:
        print(f" Temperature over 70 degrees {temp}.")
```
