```python
# How many grades in the list
count = len(grades)
print(f'Number of count: {count}')
```

```python
# Total count
total = sum(grades)
print(f'Total count: {total}')

# Average count
average =  total / count
print(f'Average: {average:.2f}')
```

```python
# Highest and Lowest
print(f'Highest: {max(grades)}')
print(f'Lowest: {min(grades)}')
```

```python
# Sort in ascending order
grades.sort()
print(grades)
```

```python
# Reverse in descending order
grades.reverse()
print(grades)

#This is a real grade book!
```
