```python
# Tasks
tasks = ['email', 'meeting', 'lunch']
```

```python
# Append report
tasks.append('report')
print(tasks)
```

```python
# Insert in certain postion
tasks.insert(1,'coffee')
print(tasks)
```

```python
# Remove last item
tasks.remove('lunch')
print(tasks)
```

```python
# Remove last items
last = tasks.pop()
print(last)
print(tasks)
```
