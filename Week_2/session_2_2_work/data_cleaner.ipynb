```python
# Given messy sensor data:
data = [10, -999, 25, 30, -999, 15, 40, -999, 35, 20]
```

```python
# (-999 means sensor error -remove it!)

data_filter = [d for d in data if d > 0]
print(f"Data: {data}")
print(f"Filter Data: {data_filter}")
```

```python
# How many numbers
count = len(data_filter)
print(f'Number of count: {count}')
```

```python
# Total count
total = sum(data_filter)
print(f'Total count: {total}')
```

```python
# Filter unwanted # Average count
average = total / count
print(f'Average: {average:.2f}')
```

```python
# Max and Min
max_val = max(data_filter)
min_val = min(data_filter)
print(f'Max: {max_val}')
print(f'Min: {min_val}')
```

```python
# Normalize Data
# Apply normalization to each element using list comprehension

normalized = [(x - min_val) / (max_val - min_val) for x in data_filter]
normalized_rounded = [round(val, 2) for val in normalized]
print(f"Filter Data: {data_filter}\n")
print(f"Normalize: {normalized_rounded}")
```
