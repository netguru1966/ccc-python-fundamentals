```python
# Two parallel lists
names = ['Alice', 'Bob', 'Carol', 'David']
scores = [95, 87, 92, 78]
```

```python
# Combine with zip()
for name, score in zip(names, scores):
    print(f'{name}: {score}')
```

```python
# Create dictionary from two lists
grade_dict = dict(zip(names, scores))
print(grade_dict)
```
