```python
# Raw sensor data (Fahrenheit)
temps_f= [32, 68, 98.6, 212, 72, 86, 50, 95]
```

```python
# Convert to Celsius, round

for temp in temps_f:
    if temp > 0:  # Filter out errors
        # Convert each temperature individually inside the loop
        temp_c = round((temp - 32) * 5/9, 1)
        print(f"Fahrenheit {temp}F    Celsius: {temp_c}C")
        
        # Check if the Celsius temperature is over 20C
        if temp_c > 20:
            print(f"Fahrenheit {temp}F    Celsius over 20C: {temp_c}C")
        # Check if the Celsius temperature is over 20C
        if temp_c > 25:
            print(f" HOT!!! Fahrenheit {temp}F    Celsius over 20C: {temp_c}C")
        else:
            print(f" COLD!!! Fahrenheit {temp}F    Celsius over 20C: {temp_c}C")
```
