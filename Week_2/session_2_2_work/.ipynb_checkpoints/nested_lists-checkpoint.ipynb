```python
# Student grades: [student][subject]
# Columns: Math, English, Science

grades = [
    [95, 87, 92], # Alice
    [88, 90, 85], # Bob
    [92, 88, 94], # Carol
    [78, 85, 80]# David
]
avgs = []
```

```python
# Access specific grade
print(grades[0][0]) # 95 (Alice's Math)
print(grades[1][2]) # 85 (Bob's Science)
```

```python
print(grades[0])
```

```python
for student_grades in grades:
    # Calculate average for current student and append to avgs list
    avg = sum(student_grades)/len(student_grades)
    avgs.append(avg)

# Print each student's average
for i, avg in enumerate(avgs):
    print(f'Student {i+1} Average: {avg:.1f}')

# Or if you want to print the entire list
print(f'Averages: {[round(avg, 1) for avg in avgs]}')
```
