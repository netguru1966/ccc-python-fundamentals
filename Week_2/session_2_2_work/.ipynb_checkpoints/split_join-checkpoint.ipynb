```python
# Join words with spaces
words = ['I', 'love', 'Python']
sentence = ' '.join(words)
print(sentence) # 'I love Python'
```

```python
# Join with different separator
words = ['apple', 'banana', 'cherry']
csv = ','.join(words)
print(csv) # 'apple,banana,cherry'
```

```python
# Join with different separator
words = ['apple', 'banana', 'cherry']
csv = '\t'.join(words)
print(csv) # 'apple,banana,cherry'
```

```python
# Join with newlines
lines = ['Line 1', 'Line 2', 'Line 3']
text = '\n'.join(lines)
print(text)
```
