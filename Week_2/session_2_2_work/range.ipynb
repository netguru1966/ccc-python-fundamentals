```python
# First 10 squares
squares = [i**2 for i in range(10)]
print(squares)
```

```python
# Even numbers 0 to 20
evens = [i for i in range(21) if i % 2 == 0]
print(evens)
```

```python
# Multiples of 5 up to 50
mult_5 = [i for i in range(0, 51, 5)]
print(mult_5)
```

```python
# Multiples of 7 up to 84
mult_7 = [i for i in range(0, 85, 7)]
print(mult_7)
```

```python
# Multiples of 2 up to 26
mult_2 = [i for i in range(0, 27, 2)]
print(mult_2)
```

```python
# Generate test data instantly!
test_data = [i * 0.1 for i in range(100)]
print(test_data)

# Option 2: Format each element in the list to 2 decimal places|
formatted_data = [f'{x:.1f}' for x in test_data]
print(f'\n Formatted Test Data: {formatted_data}')
```
