```python
# WRONG! (This is a trap!)

original = [1, 2, 3, 4, 5]

reference = original # Not a copy!
reference.append(6)
print(original) # - CHANGED!

print(reference)
```

```python
# RIGHT! Make actual copy
original = [1, 2, 3, 4, 5]

vanay_copy = original.copy() # Real copy
vanay_copy.append(6)
print(original) # - Unchanged!
print(vanay_copy) # Copy
```

```python
# Other ways to copy:
# copy = list(original)
# copy = original[:]
# All three create real copies!
```

```python
# RIGHT! Make actual copy
original = [1, 2, 3, 4, 5, 6]

vanay_copy = list(original) # Real copy
vanay_copy.append(7)
print(original) # - Unchanged!
print(vanay_copy) # Copy
```

```python
# RIGHT! Make actual copy
original = [10, 20, 30, 40, 50, 60]

vanay_copy = original[:] # Real copy
vanay_copy.append(70)
print(original) # - Unchanged!
print(vanay_copy) # Copy
```
