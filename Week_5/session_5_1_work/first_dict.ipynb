### First Dictionary

```python
# Create a person dictionary

person = {
    'name': 'Alice',
    'age': 25,
    'city': 'Boston',
    'job': 'Data Scientist'
}
```

```python
# Access values by key
print(person['name'])  # Alice
print(person['age'])   # 25
print(person['city'])  # Boston
```

```python
# Much clearer than:
# person_list = ['Alice', 25, 'Boston', 'Data Scientist']
# print(person_list[0])  # What is this?
# Self-documenting code!
```

 ### What is this? - List
