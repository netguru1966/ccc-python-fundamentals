### Creating Dictionaries - Different Ways

```python
# Method 1: Literal (most common)

student = {'name': 'Bob', 'grade': 85}
```

```python
# Method 2: Empty then add

student = {}
student['name'] = 'Bob'
student['grade'] = 85
```

```python
# Method 3: dict() function

student = dict(name='Bob', grade=85)
```

```python
# All create the same dictionary!
# Use Method 1 most of the time
```

```python
# Multi-line for readability:

student = {
    'name': 'Bob',
    'grade': 85,
    'major': 'CS'
}
```

```python
# Access values by key
print(student['name'],end=" ")  # Bob
print(student['grade'],end=" ") # 85
print(student['major'],end=" ") # CS
```
