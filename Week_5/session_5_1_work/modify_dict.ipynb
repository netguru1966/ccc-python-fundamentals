### Modifying Dictionary Examples

```python
student = {'name': 'Bob', 'age': 20}
```

```python
# Add new field
student['major'] = 'Computer Science'

print(student)
# {'name': 'Bob', 'age': 20, 'major': 'Computer Science'}
```

```python
# Update existing field

student['age'] = 21
print(student['age'])  # 21
```

```python
print(student)
```

```python
# Delete a field
del student['age']

print(student)
# {'name': 'Bob', 'major': 'Computer Science'}
```
