### Your First Nested Dictionary

```python
# Students with detailed info

students = {
    'Alice': {
        'age': 20,
        'major': 'CS',
        'gpa': 3.8
    },
    'Bob': {
        'age': 21,
        'major': 'Math',
        'gpa': 3.6
    }
}
```

```python
# Access nested data

print(students['Alice'])            # Whole dict

# {'age': 20, 'major': 'CS', 'gpa': 3.8}
```

```python
print(students['Alice']['major'])   # Specific value
# CS
# Two levels of access!
```

```python
print(students['Bob']['gpa']) 
```
