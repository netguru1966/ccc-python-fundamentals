### Nested Dictionary Access Examples

```python
company = {
    'engineering': {
        'employees': 50,
        'manager': 'Alice',
        'budget': 5000000
    },
    'marketing': {
        'employees': 20,
        'manager': 'Bob',
        'budget': 2000000
    }
}
```

```python
# Access department info

eng_manager = company['engineering']['manager']
print(eng_manager)    # Alice
```

```python
# Calculate total employees
total = company['engineering']['employees'] + \
        company['marketing']['employees']
print(f'Total Employees: {total}')   # 70
```

```python
# Calculate total budget
total = company['engineering']['budget'] + \
        company['marketing']['budget']
print(f'Total Budget: {total}')   # 7000000
```
