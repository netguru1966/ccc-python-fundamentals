### Build nested structure:

```python
school = {
    'math_dept': {
        'teachers': ['Dr. Smith', 'Prof. Jones'],
        'students': 550,
        'budget': 50000
    },
    'science_dept': {
        'teachers': ['Dr. Henry', 'Prof. Baker'],
        'students': 375,
        'budget': 180000
    },
    'english_dept': { ... }
}
```

```python
# Then:
# 1. Add science_dept
# 2. Calculate total students
# 3. List all teachers (from all depts)
# 4. Find dept with most students
# 5. Calculate total budget

# Real organizational data!
```

```python
# Calculate total students
total = school['math_dept']['students'] + \
        school['science_dept']['students']
print(f'Total Students: {total}') 
```

```python
print(school['math_dept']['teachers'])
print(school['science_dept']['teachers'])
```

```python
# Find department with maximum students

# Create a dictionary to store department names and their student counts

departments = {
    'math_dept': school['math_dept']['students'],
    'science_dept': school['science_dept']['students']
}

# Find the department with maximum students
highest_dept = max(departments, key=departments.get)
highest_count = departments[highest_dept]

print(f"Department with max students: {highest_dept} with {highest_count} students")
```

```python
# Calculate total budget
total = school['math_dept']['budget'] + \
        school['science_dept']['budget']
print(f'Total Budget: {total}') 
```
