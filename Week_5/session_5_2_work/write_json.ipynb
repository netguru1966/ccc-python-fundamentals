### Write JSON File Example

```python
import json
```

```python
# Configuration data

config = {
    'model_type': 'neural_network',
    'learning_rate': 0.001,
    'epochs': 150,
    'layers': [128, 64, 32]
}
```

```python
# Write to file

with open('config.json', 'w') as file:
    json.dump(config, file, indent=2)
```
